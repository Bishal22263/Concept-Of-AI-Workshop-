```python
# Problem - 3: Array Operations:

import numpy as np

x = np.array([[1,2],
              [3,5]])
y = np.array([[5,6],
              [7,8]])

v = np.array([9,10])
w = np.array([11,12])

# 1

print("Add of x and y: ",np.add(x,y))
print("Add  of v and w: ",np.add(v,w))

# 2

print("Subtract of y from x: ",np.subtract(x,y))
print("Subtract of  w from v: ",np.subtract(v,w))

# 3

print("Multiplying x by 3: ",np.multiply(x,3))
print("Multiplying y by 2: ",np.multiply(y,2))
print("Multiplying v by 5: ",np.multiply(v,5))
print("Multiplying w by 4: ",np.multiply(w,4))

# 4
print("Square of x =\n", x ** 2)
print("Square of y =\n", y ** 2)
print("Square of v =\n", v ** 2)
print("Square of w =\n", w ** 2)

# 5
print("The dot product between v and w is ", np.dot(v,w))
print("The dot product between x and v is ", np.dot(x,v))
print("The dot product between x and y is ", np.dot(x,y))

# 6

np.concatenate((x, v))
# all the input arrays must have same number of dimensions,  but the array at index 0 has 2 dimension(s) and the array at index 1 has 1 dimension(s)



```
